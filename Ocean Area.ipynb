```python
import math

def calculate_1degree_ocean_area(latitude_deg):
    # Constants for Earth's oblate spheroid
    equatorial_radius = 6378137.0  # in meters
    polar_radius = 6356752.3142  # in meters

    # Convert latitude to radians
    latitude_rad = math.radians(latitude_deg)

    # Calculate the length of 1 degree of longitude at the given latitude
    degree_lon_length = 2 * math.pi * equatorial_radius * math.cos(latitude_rad) / 360.0

    # Calculate the area of the ocean strip at the given latitude
    ocean_area = degree_lon_length * polar_radius * math.radians(1)

    return ocean_area

# Calculate ocean areas for each 1° latitude from -90° to 90°
for latitude in range(-90, 91):
    ocean_area = calculate_1degree_ocean_area(latitude)
    print(f"At latitude {latitude}°: {ocean_area:.2e} square meters")
```

```python
data_file_path = 'C:\\Users\\yxeng\\OneDrive - University of Tasmania\\Life Part 2\\Thesis\\Chapter 5 - Phylogeny Modeling\\Figures\\Ocean area\\Latitudes_1degree.txt'
```

```python
data = []
with open(data_file_path, 'r') as file:
    next(file)  # Skip the header line
    for line in file:
        if line.strip():  # Skip empty lines
            parts = line.strip().split('\t')  # Split using tabs
            lower, upper = float(parts[0]), float(parts[1])
            region = f"Region {int(lower)}"  # Generating a default region name based on lower value
            data.append((lower, upper, region))
```

```python
# Create a new vector layer for clipping polygons
clipping_layer = QgsVectorLayer("Polygon?crs=epsg:4326", "clipping_polygons", "memory")
dataProvider = clipping_layer.dataProvider()
dataProvider.addAttributes([QgsField("region", QVariant.String)])

# Start edit session on the clipping layer
clipping_layer.startEditing()

# Loop through data and create clipping polygons
for range_info in data:
    polygon = QgsFeature()
    pts = [
        QgsPointXY(-180, range_info[0]),
        QgsPointXY(180, range_info[0]),
        QgsPointXY(180, range_info[1]),
        QgsPointXY(-180, range_info[1])
    ]
    polygon.setGeometry(QgsGeometry.fromPolygonXY([pts]))
    polygon.setAttributes([range_info[2]])
    dataProvider.addFeatures([polygon])

# Commit changes and add the clipping layer to the map
clipping_layer.commitChanges()
QgsProject.instance().addMapLayer(clipping_layer)
```
